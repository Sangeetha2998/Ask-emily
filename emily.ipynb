```python
with open('interpreter.yml','w') as f:
  f.write('''
  language: "en"

  pipeline:
  - name: "nlp_mitie"
    model: "MITIE-models/english/total_word_feature_extractor.dat"
  - name: "tokenizer_mitie"
  - name: "ner_mitie"
  - name: "ner_synonyms"
  - name: "intent_entity_featurizer_regex"
  - name: "intent_featurizer_mitie"
  - name: "intent_classifier_sklearn"
  ''')
```

```python
with open('interpreter_dataset.md','w') as f:
  f.write('''

  ## intent:intent_bye
  - Bye
  - Goodbye
  - See you later
  - Gotta go
  - See you
  - I'm off

  ## intent:intent_greet
  - Hi
  - Hey
  - Hello
  - Good morning
  - Hi there
  - Greetings

  ## intent:intent_thank
  - Thanks
  - Thank you
  - Cheers
  - Thanks a lot
  - Cool, thank you

  ## intent:intent_yes
  - Yes
  - Yeah
  - Sure
  - Absolutely
  - Yep

  ## intent:intent_no
  - No
  - No way
  - Nope
  - Absolutely not
  - I don't think so

  ## intent:intent_name
  - My name is [Alice](name)
  - I'm [Lucy](name)
  - I am [Richard](name)
  - Call me [Sally](name)

  ## intent:intent_question
  - I have a question about [campgrounds](topic)
  - Can I ask you something about [routes](topic)?
  - What's the best route for my [RV](topic)?
  - Can you help me find an [RV technician](topic)?
  - Do you know where I can find diesel nearby?
  - Can you help me understand the [Roamer Rewards](topic)?

  ## intent:intent_find_route
  - I need help finding a route for my [RV](rig_type)
  - Can you create a route for my [RV](rig_type)?
  - What's the best route for my [truck](rig_type)?
  - Show me a route with fuel stations for my [RV](rig_type)
  - Can you plan a route avoiding low bridges for my [rig](rig_type)?

  ## intent:intent_check_campgrounds
  - Show me campgrounds nearby
  - Are there any campgrounds along my route?
  - What are the options for camping around here?
  - Find campgrounds close to my destination
  - I need recommendations for campgrounds

  ## intent:intent_fuel_station_check
  - Where can I find diesel on my route?
  - Show me nearby fuel stations
  - Is there a gas station with diesel near me?
  - Can you show fuel prices for stations along my route?
  - I need a gas station with diesel nearby

  ## intent:intent_locate_technician
  - I need an RV technician
  - Can you help me find an RV repair service?
  - Show me nearby RV repair services
  - Find an RV technician close to my location
  - I need help with RV repairs

  ## intent:intent_show_rewards
  - Show me my rewards
  - What Roamer Rewards do I have?
  - Do I have any discounts on this trip?
  - Are there any rewards I can use?
  - Show me available Roamer Rewards

  ''')
```
